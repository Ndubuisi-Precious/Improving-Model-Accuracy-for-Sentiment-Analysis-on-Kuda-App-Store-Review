```python
import neattext.functions as nfx
import re
import string
from nltk import word_tokenize
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
def data_processing(text):
    text = text.lower()
    text = re.sub(r'https\S+|WWW\S+https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\@W+|\#','',text)
    text = re.sub(r'[^\W\S]','',text)
    text = re.sub(r'""','',text)
    text = re.sub(r"'",'',text)
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words]
    return " ".join(filtered_text)
```

```python
kuda_df['cleaned_review'] = kuda_df['review'].apply(data_processing)

kuda_df.head(4)
```

```python
#dir(nfx)
```

```python
#### cleaning text remove multiple spaces, emojies, and special characters
kuda_df['cleaned_review'] = kuda_df['cleaned_review'].apply(nfx.remove_stopwords)
kuda_df['cleaned_review'] = kuda_df['cleaned_review'].apply(nfx.remove_multiple_spaces)
kuda_df['cleaned_review'] = kuda_df['cleaned_review'].apply(nfx.remove_puncts)
kuda_df['cleaned_review'] = kuda_df['cleaned_review'].apply(nfx.remove_punctuations)
kuda_df['cleaned_review'] = kuda_df['cleaned_review'].apply(nfx.remove_special_characters)
kuda_df['cleaned_review'] = kuda_df['cleaned_review'].apply(nfx.remove_emojis)

kuda_df.head(4)
```

```python
def get_sentiment(rating):
    if rating < 3:
        return 'Negative'
    if rating == 3: 
        return'Average'
    if rating > 4:
        return 'Positive'
```

```python
kuda_df['Sentiment'] = kuda_df['rating'].apply(get_sentiment)
kuda_df.head(5)
```

## Feature Extraction

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from imblearn.over_sampling import SMOTE
```

#### Model Training and Evaluation
+ I will implement and train multiple models using the prepared features and rating
- Logistic regression
- Naive Bayes 
- K-Nearest Neighbors (KNN)
- SVC

```python
#let's split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(kuda_df['cleaned_review'], kuda_df['rating'],test_size=0.2, random_state=42)
```

```python
# Let's vectorize the text data

vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
#Logistic Regression Model
lr_model = LogisticRegression()
lr_model.fit(X_train_vec, y_train)
lr_prediction = lr_model.predict(X_test_vec)
lr_accuracy = accuracy_score(lr_prediction, y_test)
lr_cm = confusion_matrix(lr_prediction, y_test)
print('Regression Accuracy:', lr_accuracy)
print('\n')
print('Test accuracy: {:.2f}%'.format(lr_accuracy*100))
print('\n')
print(classification_report(y_test, lr_prediction))
print('Regression confusion Matrix:', lr_cm)
```

```python
# Naive Bayes model
nb_model = MultinomialNB()
nb_model.fit(X_train_vec, y_train)
nb_prediction = nb_model.predict(X_test_vec)
nb_accuracy = accuracy_score(nb_prediction, y_test)
nb_cm = confusion_matrix(nb_prediction, y_test)
print('Bayes Accuracy:', nb_accuracy)
print('Test accuracy: {:.2f}%'.format(nb_accuracy*100))
print('\n')
print(classification_report(y_test, nb_prediction))
print('\n')
print('Regression confusion Matrix:', nb_cm)
```

```python
# Support Vector Machine (SVM) model

svm_model = SVC()
svm_model.fit(X_train_vec, y_train)
svm_prediction = svm_model.predict(X_test_vec)
svm_accuracy = accuracy_score(svm_prediction, y_test)
svm_cm = confusion_matrix(svm_prediction, y_test)
print('SVM Accuracy:', svm_accuracy)
print('Test accuracy: {:.2f}%'.format(svm_accuracy*100))
print('\n')
print(classification_report(y_test, svm_prediction))
print('\n')
print('Regression confusion Matrix:', svm_cm)
```

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_train_vec, y_train)
knn_prediction =knn_model.predict(X_test_vec)
knn_accuracy = accuracy_score(y_test, knn_prediction)
knn_cm = confusion_matrix(y_test, knn_prediction)
print('SVM Accuracy:', knn_accuracy)
print('Test accuracy: {:.2f}%'.format(knn_accuracy*100))
print('\n')
print(classification_report(y_test, knn_prediction))
print('\n')
print('Regression confusion Matrix:', knn_cm)
```
